```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Zuhaib haider\Downloads\train.csv.zip")
df.head()
```

```python
df.info()
```

```python
df['Postal Code'].isnull().sum()
```

```python
df = df.dropna(subset=['Postal Code'])
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df.dtypes
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
#Delivery Time
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df[['Order Date' , 'Ship Date' , 'Delivery Time']].head()
```

```python
#Average delivery time
df['Delivery Time'].mean()
```

```python
#Find Max and Min delivery Time
df['Delivery Time'].max()
df['Delivery Time'].min()
```

```python
#Which Mode takes the least time to deliver
df.groupby('Ship Mode')['Delivery Time'].mean()
```

```python
#How much time it take on an average to deliver in each region
df.groupby ('Region')['Delivery Time'].mean()
```
